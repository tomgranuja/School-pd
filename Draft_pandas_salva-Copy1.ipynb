```python
def df_maker(data_file):
    table=pd.read_table(data_file, sep=',', encoding='latin1')
    
    date = file_name_to_date(data_file)
    date_col = [date for i in range(table.shape[0])] #future date column
    
    ## Splitting 'horario' column into two columns, 'dia' and 'hora', dtypes 'category' and 'int' respectively ##
    day_time = pd.DataFrame(table.horario.str.split(' ',1).tolist(),  columns=['dia', 'hora'])
    day_time.dia= day_time.dia.astype('category', categories=['Lun', 'Mar','Mie', 'Jue', 'Vie', 'Sab'], ordered=True)
    day_time.hora = day_time.hora.str.replace(':', '').astype(int)
    
    table.loc[:,'date'] = pd.Series(date_col)#Adding a new Timestamp column named 'date' 
    table[['dia', 'hora']] = day_time
    # table[['dia', 'hora']] = day_time.loc[:,['dia','hora']] ##works too

    table.drop('horario', axis=1, inplace=True) #Deletes 'horario' column
    table.set_index('date', inplace=True) #making the index (useful??)
    
    return table


test_table = df_maker(file1)
test_table
```
